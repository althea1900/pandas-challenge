```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
#school_data_complete.dtypes
#school_data.dtypes
student_data.dtypes
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Calculate the total number of schools
total_school = len(school_data_complete["school_name"].unique())

# Calculate the total number of students
total_students = len(school_data_complete["student_name"])

# Calculate the total budget
total_budget = sum(school_data_complete["budget"].unique())

# Calculate the average math score
avg_math_score = school_data_complete["math_score"].mean()

# Calculate the average reading score
avg_reading_score = school_data_complete["reading_score"].mean()

# Calculate the percentage of students with a passing math score (70 or greater)
# students_passing_math = (school_data_complete[school_data_complete["math_score"]>69]
#                          ["math_score"].count()/total_students) * 100

percentage_passing_math = ((school_data_complete["math_score"]>69).sum() / school_data_complete["math_score"].count())*100

# Calculate the percentage of students with a passing reading score (70 or greater)
percentage_passing_reading = (school_data_complete[school_data_complete["reading_score"]>69]
                         ["reading_score"].count()/total_students) * 100

# Calculate the percentage of students who passed math and reading (% Overall Passing)
percentage_passing_both = (percentage_passing_math * percentage_passing_reading)/ 100

# percentage_passing_both = [school_data_complete['reading_score']> 69 & school_data_complete['math_score'] > 69]
 
#total_school, total_students, total_budget, avg_math_score, avg_reading_score, percentage_passing_math, percentage_passing_reading, percentage_passing_both
    
# Create a dataframe to hold the above results

district_summary_df = pd.DataFrame({"Total Schools":[total_school],
                               "Total Students":[total_students],
                               "Total Budget" : [total_budget],
                               "Average Math Score":[avg_math_score],
                               "Average Reading Score":[avg_reading_score],
                               "% Passing Math":[percentage_passing_math],
                               "% Passing Reading": [percentage_passing_reading],
                               "% Overall Passing":[percentage_passing_both]})

district_summary_df

# Optional: give the displayed data cleaner formatting


```

```python
# Optional: give the displayed data cleaner formatting
district_summary_df = pd.DataFrame({"Total Schools":[total_school],
                               "Total Students":[total_students],
                               "Total Budget" :"${:,.2f}".format(total_budget),
                               "Average Math Score":"{:,.2f}".format(avg_math_score),
                               "Average Reading Score":"{:,.2f}".format(avg_reading_score),
                               "% Passing Math":"{:.2f}%".format(percentage_passing_math),
                               "% Passing Reading":"{:.2f}%".format(percentage_passing_reading),
                               "% Overall Passing":"{:.2f}%".format(percentage_passing_both)})

district_summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Create an overview table that summarizes key metrics about each school, including:
#  - School Name
#school_groups = school_data_complete.groupby(["school_name"], as_index=False)
school_groups = school_data_complete.set_index("school_name").groupby(["school_name"])
#school_groups.head()

# #  - Total Students
total_students = school_groups["Student ID"].count()
# total_students

#  - School Type
school_types = school_data.set_index("school_name")["type"]
#school_types.head()

#  - Total School Budget
school_budgets = school_groups["budget"].first()
# school_budgets.head()

#  - Per Student Budget
per_student_budget = school_budgets/total_students
#per_student_budget.head()

#  - Average Math Score
average_math_scores = school_groups["math_score"].mean()
# average_math_scores.head()

#  - Average Reading Score
average_reading_scores = school_groups["reading_score"].mean()
#average_reading_scores.head()

#  - % Passing Math
passing_math_group = school_data_complete[school_data_complete["math_score"]>69].groupby(["school_name"]).size()
percent_passing_math = (passing_math_group / total_students) * 100
#percent_passing_math.head()

#  - % Passing Reading
passing_reading_group = school_data_complete[school_data_complete["reading_score"]>69].groupby(["school_name"]).size()
percent_passing_reading = (passing_reading_group / total_students) * 100
#percent_passing_reading.head()

#  - % Overall Passing (The percentage of students that passed math and reading.)
percent_overall_passing = (percent_passing_reading * percent_passing_math) / 100
#percent_overall_passing.head()

#     Create a dataframe to hold the above results
school_summary = pd.DataFrame({"School Type" : school_types,
                               "Total Students" : total_students,
                               "Total School Budget" : school_budgets,
                               "Per Student Budget" : per_student_budget,
                               "Average Math Score" : average_math_scores,
                               "Average Reading Score" : average_reading_scores,
                               "% Passing Math": percent_passing_math,
                               "% Passing Reading" : percent_passing_reading,
                               "% Overall Passing" : percent_overall_passing})
 
school_summary = school_summary [["School Type", 
                               "Total Students", 
                               "Total School Budget",
                               "Per Student Budget",
                               "Average Math Score",
                               "Average Reading Score",
                               "% Passing Math",
                               "% Passing Reading",
                               "% Overall Passing"]]

school_summary['Total Students'] = school_summary['Total Students'].map("{:,}".format)
school_summary['Total School Budget'] = school_summary['Total School Budget'].map("${:,}".format)
school_summary['Per Student Budget'] = school_summary['Per Student Budget'].map("${:.2f}".format)
school_summary['Average Math Score'] = school_summary['Average Math Score'].map("{:,.2f}".format)
school_summary['Average Reading Score'] = school_summary['Average Reading Score'].map("{:,.2f}".format)
school_summary['% Passing Math'] = school_summary['% Passing Math'].map("{:.2f}%".format)
school_summary['% Passing Reading'] = school_summary['% Passing Reading'].map("{:.2f}%".format)
school_summary['% Overall Passing'] = school_summary['% Overall Passing'].map("{:.2f}%".format)

# school_summary_df = pd.DataFrame({"School Type":[school_types],
#                                "Total Students":[total_students],
#                                "Total School Budget":[total_students],
#                                "Per Student Budget":[per_student_budget],
#                                "Average Math Score":[average_math_scores],
#                                "Average Reading Score":[average_reading_scores],
#                                "% Passing Math":[percent_passing_math],
#                                "% Passing Reading":[percent_passing_reading],
#                                "% Overall Passing":[percent_overall_passing]})


school_summary
```
